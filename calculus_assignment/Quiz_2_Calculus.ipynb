# Task 1: Write a dynamic function to find the derivative of any function f(x)

```python
import sympy as sp
import matplotlib.pyplot as plt
from sympy import symbols, diff, plot
import numpy as np
```

```python
#TO DO: Create a function that does a derivative for any function
def derivative(f, var='x'):
    f = sp.sympify(f)
    return sp.diff(f, var)
```

# Task 2: Test the derivative function written with a quadratic equation of your choice
*NB: Must have atleast 2 minimas and atleast 2 maximas*

```python
# Define the quadratic function
x = sp.Symbol('x')
f = x**2 - 4*x + 5
```

```python
# Test the derivative function
print("Derivative:", derivative(f))
```

#Task 3: Plot a graph of the quadratic Equation

```python
#Create a visualization of the quadratic equation
x_vals = np.linspace(-5, 5, 100)
y_vals = [f.evalf(subs={x:i}) for i in x_vals]
fig, ax = plt.subplots()
ax.plot(x_vals, y_vals)
ax.set(xlabel='x', ylabel='f(x)',
       title='Graph of quadratic function')
```

```python
# Find critical points
crit_points = sp.solve(derivative(f))
print("Critical points:", crit_points)
```

# Task 4: Create separate array of Minimas and maximas

```python
arr_minimas = []
arr_maximas = []

for cp in crit_points:
    second_deriv = derivative(derivative(f), x).evalf(subs={x:cp})
    if second_deriv > 0:
        arr_minimas.append((cp, f.evalf(subs={x:cp})))
    elif second_deriv < 0:
        arr_maximas.append((cp, f.evalf(subs={x:cp})))

print(f'the array of minimas {arr_minimas} and the array of maximas is {arr_maximas}')
```

#Finally, What is the global Minima and the Global maxima _ Plot this so that I can see

```python
#Task 5: Find global min/max
if len(arr_minimas) > 0:
    Global_minima = min(arr_minimas, key=lambda x: x[1])
else:
    Global_minima = None

if len(arr_maximas) > 0:
    Global_maxima = max(arr_maximas, key=lambda x: x[1])
else:
    Global_maxima = None

print(f'The global maxima is {Global_maxima}')
print(f'The global minima is {Global_minima}')
```

```python
# Creates a figure and axes
fig, ax = plt.subplots()

# Plots the critical points
if Global_maxima is not None:
    ax.plot(Global_maxima[0], Global_maxima[1], "go", label="Global Maxima")
if Global_minima is not None:
    ax.plot(Global_minima[0], Global_minima[1], "ro", label="Global Minima")

# Plots the quadratic equation
x_vals = np.linspace(-5, 5, 100)
y_vals = [f.evalf(subs={x:i}) for i in x_vals]
ax.plot(x_vals, y_vals, label="Quadratic Equation")
ax.set(xlabel='x', ylabel='f(x)',
       title='Graph of quadratic function')

# Adds legend and displays the graph
ax.legend()
plt.show()
```
